## Pra Pemrosesan Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df_baca = pd.read_csv('/content/sample_data/Minat baca.csv')
df_baca
```

```python
# cek data null
is_na = df_baca.isna().all(axis=0)
is_na
```

isna() digunakan untuk mengecek apakah ada nilai NaN pada dataframe.

```python
# cek data duplikat
duplicate = df_baca.drop_duplicates(inplace = True)
print(duplicate)
```

```python
df_baca.describe()
```

describe() digunakan untuk menampilkan deskriptif statistik data. Hanya kolom yang bertipe numerik yang akan ditampilkan statistiknya.

```python
df_baca.info()
```

info() digunakan untuk menampilkan informasi detail tentang dataframe, seperti jumlah baris data, nama-nama kolom berserta jumlah data dan tipe datanya, dan sebagainya.

```python
df_baca.isnull().sum()
```

```python
df_duplicated_free = df_baca.drop_duplicates()
dp = df_duplicated_free.groupby(['Nama'])['Nama'].count()
print([dp])
```

## Reduksi Data

```python
corr = df_baca.corr()
plt.figure(figsize=(13,10))
sns.heatmap(corr, annot = True, cmap = plt.cm.Reds)
plt.show()
```

## Teknik Plotting Menggunakan Scatter Plot

```python
x = df_baca.IPK
y = df_baca.Universitas
```

```python
plt.scatter(x,y, s = 50, )
plt.title('IPK Mahasiswa Berdasarkan Universitas')
plt.xlabel('IPK Mahasiswa')
plt.ylabel('Universitas')

plt.show()
```
